```python
from datasets import load_dataset, Image
import os

DATA_DIR = r"D:\Confocal_imaging_nuclei_tif\MIST_Fused_Images\ROIs"

ds = load_dataset("imagefolder", data_dir=DATA_DIR, split="train", keep_in_memory=False)

# (Optional) remove a dummy label if present
if "label" in ds.column_names:
    ds = ds.remove_columns("label")

# Add filename
ds = ds.cast_column("image", Image(decode=False))
def add_filename(ex): return {"filename": os.path.basename(ex["image"]["path"])}
ds = ds.map(add_filename, batched=False)
ds = ds.cast_column("image", Image())

print(ds)           # quick check
print(ds.features)  # should show {'image': Image(...), 'filename': Value('string')}
```

```python
REPO_ID = "RussellBarkley/NucleusNet"  # or "RussellBarkley/NucleusNet-v1" etc.
ds.push_to_hub(REPO_ID, private=True, max_shard_size="1GB")
```
