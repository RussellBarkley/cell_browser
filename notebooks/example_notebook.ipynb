```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
```

Jupyter notebooks can have a mixture of markdown and code cells, which will be rendered in the deployed site.
E.g. here is a static plot -- which depends on a random seed, thus would be different every time it's executed.

```python
x = np.linspace(0,4*np.pi,256)
y = np.sin(x)
y += np.random.rand(256)/2

plt.plot(x,y)
```

```python
def wrapper_function(noise_amplitude):
    y = np.sin(x)
    y += np.random.rand(256) * noise_amplitude
    plt.plot(x,y)
    return None
```

```python
#| label: app:example_widget
# Exampel Widget

ipywidgets.interact(
    wrapper_function,
    noise_amplitude=(0.1,2.0,0.1)
)
```
